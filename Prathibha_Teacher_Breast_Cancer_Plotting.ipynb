```
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import matplotlib.colors
```

```
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["red","yellow","green"])
```

```
data = load_breast_cancer()
```

#Printing Feature Names, so that we can understand the features order

```
print(data.feature_names) #see the different features
print(data.feature_names.shape) #see the no of features is 30
```

#Suppose i need the mean radius and mean texture

```
print(type(data.data)) #see the type of the data is numpy array so we can easily slice
```

```
data_two_features = data.data[:,0:2] #takig all rows and 0,1 columns
print(data_two_features.shape) #prints the shape of the new array
print(data.target.shape) #data.target contains the target labels/classes (benign and malignant)
```

#Plotting

```
plt.scatter(data_two_features[:,0], data_two_features[:,1], c=data.target, cmap=my_cmap)
plt.show()
```

#Now take another two features which are adjacent

```
data_two_features = data.data[:,[3,6]] #taking 3rd and 6th feaures
```

```
print(data_two_features.shape)
```

```
plt.scatter(data_two_features[:,0], data_two_features[:,1], c=data.target, cmap=my_cmap)
plt.show() #Plotting
```
