```
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

```
R = np.random.random([100,1]) + 5
```

```
plt.plot(R)
```

```
np.mean(R)
```

```
np.std(R)
```

```
scalar = StandardScaler()
```

```
scalar.fit(R)
```

```
scalar.mean_
```

```
RT = scalar.transform(R)
```

```
plt.plot(RT,color="RED")
```

```
plt.plot(R)
plt.plot(RT,color="RED")
```
